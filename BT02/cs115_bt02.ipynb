Tuy nhiên, vì source code được viết bằng ngôn ngữ MIPS Assembly nên không tiện thể hiện ở notebook này. Xin thầy vui lòng hãy ghé qua link github chứa bài làm của em tại: https://github.com/khanh-moriaty/CS115.L11.KHTN/tree/main/BT02


Để thực thi được bài làm trên, đầu tiên cần tải phần mềm MARS để giả lập kiến trúc máy tính MIPS và assemble được source code. Tải MARS tại: https://github.com/khanh-moriaty/CS115.L11.KHTN/raw/main/BT02/Mars4_5_1_1Unofficial.jar

Sau đó, khởi động phần mềm MARS và mở file source code bằng MARS, nhấn F3 để assemble, nhấn F5 để bắt đầu chạy chương trình.
